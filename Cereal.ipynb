## Sugar and Rating of all cereals

```python
#I want to get sugar and rating
sugarAndRating = cerealDataFrame[['name','sugars', 'rating']]
sugarAndRating = sugarAndRating.set_index("name")
sugarAndRating
```

```python
sugarAndRating.sort_values(by=['rating'], ascending=False)
```

```python
highSugarDF = sugarAndRating[(sugarAndRating.sugars >= 6.92)]#.drop(['rating'])
highSugarDF = highSugarDF.drop(['sugars'], axis=1)
lowSugarDF = sugarAndRating[(sugarAndRating.sugars <= 6.91)]#.drop(['rating'])
lowSugarDF = lowSugarDF.drop(['sugars'], axis=1)
lowSugarDF
```

```python
plt.style.use('ggplot')

plt.title('High Sugar and Low Sugar Cereals')

plt.hist(highSugarDF,bins=20, alpha=0.8, label="High Sugar Cereal")
plt.hist(lowSugarDF,bins=20, alpha=0.8, label="Low Sugar Cereal")

plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.legend()

plt.show()


```
